### Mathematical Outlier Detection

### IQR

```python
def threshold_detection(temp_df,colm):
    q1, q3 = np.quantile(temp_df[colm],[0.25,0.75])
    iqr = q3 - q1
    lower_val = q1 - 1.5*iqr
    upper_val = q3 + 1.5*iqr
    print('Lower Threshold Value :',lower_val)
    print('Upper Threshold Value :',upper_val)
    return lower_val, upper_val
```

```python
#hour wise outlier detection
for col in ['beam_irradiance','diffuse_irradiance','total_irradiance']:
    print('\n Currently Working on :'+col+' \n')
    for hr in range(4,20):
        print('################## Hour ######### ',hr)
        temp_df = df.loc[df.hour==hr]
        threshold_detection(temp_df,col)
```

### Z-Score

```python
from scipy import stats 
z = np.abs(stats.zscore(df['diffuse_irradiance']))
print(z)
```

```python
print('outlier index \n',list(np.where(z > 3))[0])
```

```python
upper_threshold_value = df['diffuse_irradiance'][list(np.where(z > 3))[0]].min()
print('upper_threshold_value:',upper_threshold_value)
```
